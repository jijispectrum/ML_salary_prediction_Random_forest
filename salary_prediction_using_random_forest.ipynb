```python
# Drop Duplicates Records
df.drop_duplicates(inplace=True)

# Reset Our Index to be Orderd
df.reset_index(inplace=True, drop=True)
```

```python
df.head()
```

```python
df["Job Title"].unique()
```

```python
df = df.fillna({
    'Age': df['Age'].mean(),
    'Gender': df['Gender'].mode()[0],
    'Education Level': df['Education Level'].mode()[0],
    'Job Title': df['Job Title'].mode()[0],
    'Years of Experience': df['Years of Experience'].mean(),
    'Salary': df['Salary'].mean()
})
```

```python
df
```

```python
df.isnull().sum()
```

```python

df['Gender'] = df['Gender'].replace({'Male': 0, 'Female': 1})
```

```python
df['Education Level'] = df['Education Level'].replace({"Bachelor's": 0, "Master's": 1, "PhD":2})
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoded = LabelEncoder()
df['Job Title'] = encoded.fit_transform(df['Job Title'])
df["Job Title"].unique()
```

```python
X = df.drop(columns = ['Salary'])
y = df['Salary']
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=10, random_state=42)
model.fit(X_train, y_train)
from sklearn.metrics import mean_squared_error
# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred
# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```
